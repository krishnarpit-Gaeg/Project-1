Predict the different species of flowers on
the length of there petals and sepals
only Jupyter notebook code.


```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
# Print Target Name and Feature Name
print(iris.target_names)
print(iris.feature_names)
```

```python
import pandas as pd
data=pd.DataFrame({"sepal length(cm)":iris.data[:,0],'sepal width (cm)':iris.data[:,1], 'petal length (cm)':iris.data[:,2], 
              'petal width (cm)':iris.data[:,3],'species':iris.target})
```

```python
data
```

```python
data.head()
```

```python
x = data[['sepal length(cm)','sepal width (cm)','petal length (cm)','petal width (cm)']] # Features
y = data['species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
clf.fit(x_train,y_train)
```

```python
 # 0 'setosa' , 1 'versicolor  , 2 'virginica'[prediction]
clf.predict(x_test)
```

```python
x_test
```

```python
# score
clf.score(x_test,y_test)
```

```python
# prediction
clf.predict([[3,5,4,2]])
```
